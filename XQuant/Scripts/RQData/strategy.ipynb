```python
import pandas as pd
import numpy as np
from XQuant import Config, Formatter, Processer
from init_basic_info import *
```

```python
market_cap = market_cap_2.pivot(index="date", values="market_cap_2", columns="order_book_id")
market_cap.index = pd.to_datetime(market_cap.index)
bool_neg_market_value = market_cap.ffill().apply(lambda x: x>1e9).fillna(0)
bool_neg_market_value
```

```python
eps = epsTTM.pivot(index="date", values="epsTTM", columns="order_book_id")
eps.index = pd.to_datetime(eps.index)
eps = eps.groupby(pd.Grouper(freq="Q")).last()
eps = (eps - eps.shift(4)) / eps.shift(4)
eps = (
            eps.rolling(window=9)
            .apply(lambda x: np.nanmean(x[[-1, -5, 0]]), raw=True).ffill()
        )
bool_EPS = eps.rolling(window=2).apply(lambda x: any(x > 0)).ffill()
Formatter.transform_index(bool_EPS)
bool_EPS = Formatter.expand_dataframe(bool_EPS, begin=begin_date, end=end_date).fillna(0)
bool_EPS
```

```python

per_cash_div = dividend[["order_book_id", "quarter", "dividend_cash_before_tax"]].groupby(by=["order_book_id", "quarter"]).sum().reset_index()
per_cash_div = per_cash_div.pivot(index="quarter", values="dividend_cash_before_tax", columns="order_book_id")
q_map = dict(zip(["q" + str(i) for i in range(1,5)], Config.quarter_end))
per_cash_div.index = list(map(lambda x: x[:-2] + q_map[x[-2:]], per_cash_div.index))
per_cash_div.index = pd.to_datetime(per_cash_div.index)
per_cash_div = per_cash_div.groupby(pd.Grouper(freq="Y")).sum()
per_cash_div = per_cash_div.replace({0: np.nan})
Formatter.transform_index(per_cash_div)
per_cash_div
```

```python
per_cash_div_annual = ((per_cash_div - per_cash_div.shift(1)) / per_cash_div.shift(1)).ffill()
bool_per_cash_div = per_cash_div_annual.rolling(window=3).apply(lambda x: np.nanmean(x) >= 0.05, raw=True).ffill()
bool_per_cash_div = Formatter.expand_dataframe(bool_per_cash_div, begin=begin_date, end=end_date)
bool_per_cash_div
```

```python
per_cash_div_daily = Formatter.expand_dataframe(per_cash_div, begin=begin_date, end=end_date)
per_cash_div_daily
```

```python
turnover = price[["order_book_id", "date", "total_turnover"]].pivot(index="date", values="total_turnover", columns="order_book_id")
turnover.index = pd.to_datetime(turnover.index)
bool_turnover = Processer.pandas_parallelcal(
    turnover, func=lambda x: np.nanmean(x) >= 1e7, window=6 * 21
        ).fillna(0)
bool_turnover
```

```python
close = price[["order_book_id", "date", "close"]].pivot(index="date", values="close", columns="order_book_id")
close.index = pd.to_datetime(close.index)
close
```

```python
close, per_cash_div_daily = Processer.align_dataframe([close, per_cash_div_daily], clean=False)
```

```python
DTOP = per_cash_div_daily / close
DTOP
```

```python
bool_neg_market_value.to_csv(data_dir / "bool_neg_market_value.csv")
bool_EPS.to_csv(data_dir / "bool_EPS.csv")
bool_per_cash_div.to_csv(data_dir / "bool_per_cash_div.csv")
bool_turnover.to_csv(data_dir / "bool_turnover.csv")
DTOP.to_csv(data_dir / "DTOP.csv")
```

```python
D_t = bool_neg_market_value & bool_turnover & bool_per_cash_div & bool_EPS
D_t
```
